```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("..\\Datasets\\CSV Files\\Churn-Modelling\\Churn_modelling.csv")
```

```python
# you can make 3D plots using an actual z dimension,
# but since you use images in a slideshow, you can't rotate the
# image to get a better view of the 3D model, so it is better to use
# colors to denote the depth of a point

# LIKE MAKING A SCATTER PLOT WITH USING COLOR TO DENOTE DIMENSIONS

# using pandas
# first we have to create a new column
data["gender_color"] = data["Gender"].apply(lambda x: "blue" if x == "Male" else "red")

data.plot.scatter(x = "Age", y = "CreditScore", c = "gender_color")
```

```python
# now using matplotlib, we have more control of the color and information
for g in ["Male", "Female"]:
    # choosing a color for each gender
     c = "blue" if g == "Male" else "red"

    # filtering the dataset for the specific gender
     g_data = data[data["Gender"] == g] # this should be read as "data where data.gender == current_gender"
     plt.scatter(x = g_data["Age"], y = g_data["CreditScore"], color = c)
```

```python
# now using seaborn
sns.scatterplot(x = "Age", y = "CreditScore", hue = "Gender", data = data)
```

```python
# now using a spectrum to denote depth (using seaborn)
sns.scatterplot(x = "Age", y = "CreditScore", hue = "EstimatedSalary", data = data)
```

```python
# using pandas (when you want to denote depth using a spectrum of color, pandas is preferred)
data.plot.scatter(x = "Age", y = "CreditScore", c = "EstimatedSalary")
```
