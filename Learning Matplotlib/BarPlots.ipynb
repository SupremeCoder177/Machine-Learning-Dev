```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("..\\Datasets\\CSV Files\\Churn-Modelling\\Churn_Modelling.csv")
```

```python
data["Gender"].unique()
```

```python
gender_data = data.groupby("Gender")[["EstimatedSalary", "CreditScore"]].mean().reset_index()
```

```python
gender_data
```

```python
# using pandas
gender_data.plot.bar(x = "Gender", y = "EstimatedSalary")
```

```python
# using matplotlib
plt.bar(gender_data["Gender"], gender_data["EstimatedSalary"])
```

```python
#using seaborn
sns.barplot(x = "Gender", y = "EstimatedSalary", data = gender_data)
```

```python
# now horizontal plots (using matplotlib)
plt.barh(gender_data["Gender"], gender_data["EstimatedSalary"])
```

```python
# horizontal plots (using pandas)
gender_data.plot.barh(x = "Gender", y = "EstimatedSalary")
```
