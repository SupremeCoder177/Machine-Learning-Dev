```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
churn_data = pd.read_csv("..\\Datasets\\CSV Files\\Churn-Modelling\\Churn_Modelling.csv")
```

```python
churn_data.info()
```

```python
# creating a scatter dot graph using pandas
churn_data.sample(10000).plot.scatter(x = "Age", y = "CreditScore")
```

```python
# using matplotlib
plt.scatter(x = churn_data["Age"], y = churn_data["CreditScore"])
```

```python
# using seaborn
sns.scatterplot(data = churn_data, x = "Age", y = "CreditScore")
```

```python
# grouping by age and including the CreditScore and EstimatedSalary column
churn_age_cs = churn_data.groupby("Age")[["CreditScore", "EstimatedSalary"]].mean().reset_index()
```

```python
# showing a line plot, of ags v/s creditScore (using pandas)
churn_age_cs.plot.line(x = "Age", y = "CreditScore")
```

```python
# showing a line plot of age v/s estimated salary (using pandas)
churn_age_cs.plot.line(x = "Age", y = "EstimatedSalary")
```

```python
# using matplotlib
plt.plot(churn_age_cs["Age"], churn_age_cs["CreditScore"]) # this plots a line plot of age v/s creditscore
```

```python
# now using seaborn
sns.lineplot(churn_age_cs, x = "Age", y = "CreditScore")
```
