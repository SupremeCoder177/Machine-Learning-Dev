```python
import cv2
```

```python
# loading an image
image = cv2.imread("..\\Datasets\\Images\\img2.jpeg")
```

```python
# seeing the dimensions of the image and the no. of color channels
image.shape
```

```python
# accessing the pixel at row = 0, column = 0
image[0][0]
```

```python
# this is how you show an image in a window
cv2.imshow('image', image)
cv2.waitKey(100) # waitKey waits for keyboard input and when it gets the input it closes the image window
cv2.destroyWindow('image')
```

```python
# you can also show images using matplotlib
from matplotlib import pyplot as plt
```

```python
# for some reason the colors are inverted
# and that reason is that, openCV uses BGR and matplotlib uses RGB
plt.imshow(image)
```

```python
# converting from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

```python
# converting from BGR to HSV
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV_FULL)
```

```python
plt.imshow(img_hsv)
```

```python
# converting to gray scale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray)
```
